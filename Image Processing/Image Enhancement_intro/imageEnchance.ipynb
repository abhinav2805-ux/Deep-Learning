# Chess Board

```python
zeros = np.zeros((8,8),dtype = int)
zeros
```

```python
for i in range(len(zeros)):  
    for j in range(len(zeros[i])):
        if (i + j) % 2 == 0:
            zeros[i][j] = 1           
plt.imshow(zeros)
```

# High Intensity Image(means more pixels )

```python
img = []
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append(0)
    img.append(temp)

img = np.array(img)

print(img.shape)

plt.imshow(img)
    
```

### Exported the image

```python
cv.imwrite('img.png',img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Create a random image (grayscale)
img = np.random.rand(720, 1280)  # Values between 0 and 1

# Print shape to confirm dimensions
print(img.shape)  # Expected: (720, 1280)

# Display the image using Matplotlib
plt.imshow(img, cmap='gray')  # Use grayscale colormap
plt.show()

# Convert image to uint8 (0-255 range) for OpenCV
img_uint8 = (img * 255).astype(np.uint8)

# Save the image using OpenCV
cv.imwrite('try.png', img_uint8)
```
