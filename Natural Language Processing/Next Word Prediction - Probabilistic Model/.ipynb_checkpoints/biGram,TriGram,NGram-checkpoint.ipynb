# Markov Techinque :- Where next state is depend on the previous state 

```python
import re
import numpy as np
import pandas as pd
from tqdm import tqdm
```

# Find the pairs 

```python
with open('big.txt','r') as fd:
    lines = fd.readlines()

    words = []
    for line in lines:
        words+=re.findall('\w+',line.lower())
```

```python

def get_pairs(words , n):

    n = n+1
    
    data = []
    
    for i in range(len(words)-n):
        data.append(' '.join(words[i:i+n])) # append them in join for string

    return data
```

```python
get_pairs(words,4)
```

```python
for i in range(20): # first 20 pairs
    print(words[i],words[i+1])
```

```python
data = get_pairs(words,4)
```

```python
a = np.array(data)

pair , count = np.unique(a,return_counts = True)

print('-'*30)

print('total Pairs :- ', len(data))
unique_pair = list(set(data))
print('Unique_pairs :- ',len(unique_pair))

print('-'*30)
```

```python
prob_dist = []


for i in range(len(unique_pair)):
    prob_dist.append([unique_pair[i],' '.join(unique_pair[i].split(' ')[:-1]),unique_pair[i].split(' ')[-1],count[i]]) # (pair ,input , next word on basis of highest frequency,frequency)

print(prob_dist[0])
print(len(prob_dist))
```

# Make the DataFrame

```python
# Create a DataFrame 'df' from the 'prob_dist' list with specified column names
df = pd.DataFrame(prob_dist, columns=['seq', 'inp', 'out','freq'])

# Filter the DataFrame to include only rows where the 'freq' column is greater than or equal to 5
# df = df[df['freq'] >= 5]

# Display the first few rows of the filtered DataFrame
df.head()
```

```python
df
```

```python
def predict(word):

    if len(df[df['inp'] == word]) :
        df_ = df[df['inp']==word]
        return df_.sort_values(by = 'freq').head()['out'].values
    else:
        print('sequence is not present')
```

```python
word = 'this is a beautiful'
predict(word)
```

```python

def predict_the_word(word):
    df_pred = []
    
    for i in df.values:
        if(i[0].split(' ')[0] == word): # check the pair[0] is equal to the input word or not 
            df_pred.append([i[0],i[1],i[2]])
    
    df_pred = pd.DataFrame(df_pred , columns = ['in','freq','out'])
    
    df_pred.sort_values(by = 'freq' , ascending = False , inplace=True)
    return list(df_pred.head()['out'].values)
```

```python

word = input('enter the word')
predict_the_word(word)
```

### Make the code to generate more words (Auto Generate) but predict on the basis of one word at a time 

```python
def pred_seq(seq,n):

    output = []
    
    output.append(seq)
    
    for i in range(n):
        pred = predict(seq)
        seq = ' '.join(seq.split(' ')[1:]) + ' ' +pred[0]
        output.append(pred[0])
    
    return ' '.join(output)
        
```

```python
seq = 'of the united states'

pred_seq(seq,50)
```
